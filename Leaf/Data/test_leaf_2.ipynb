# Test set Image Processing

```python
valid_data = tf.keras.utils.image_dataset_from_directory(
    'valid',                    # Path to validation folder
    label_mode='categorical',   # Multi-class (38 classes → one-hot)
    image_size=(224, 224),      # Resize images
    batch_size=32,              # 32 images per batch
    shuffle=False               # No need to shuffle validation data
)
```

#Visualising and Performing Prediction on Single image

```python
import cv2
import matplotlib.pyplot as plt

image_path = 'test\Ax.png'

img = cv2.imread(image_path)

if img is None:
    print("❌ Image not loaded. Check path!")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title('Test Image')
    plt.axis('off')
   
    plt.show()
```

## Testing Model

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(128,128))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Convert single image to a batch.
predictions = cnn.predict(input_arr)
```

```python
class_name = valid_data.class_names
```

```python
result_index = np.argmax(predictions) #Return index of max element
print(result_index)
```

```python
# Get prediction index
import numpy as np
result_index = np.argmax(predictions)

# Get class name
model_prediction = class_name[result_index]

# Get confidence
confidence = np.max(predictions)

# Display result
plt.imshow(img)
plt.title(f"Disease Name: {model_prediction}")

print(f"Disease: {model_prediction}")
print(f"Confidence: {confidence:.2f}")

plt.show()
```
