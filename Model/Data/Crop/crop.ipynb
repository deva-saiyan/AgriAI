```python
# remove outlier
def iqr(df, col):
    df = df.copy()

    for i in col:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR   # corrected

        df[i] = df[i].clip(lower, upper)

    return df

df =iqr(df , num_col)
print("Outliers handled using IQR method".upper())
```

# ANN Model

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential # model
from keras.layers import *
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
# Early Stoping
from tensorflow.keras.callbacks import EarlyStopping

callback = EarlyStopping(
    monitor='val_loss',
    min_delta= 0.01,
    patience=5,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=True,
    start_from_epoch=0

)
```

```python
ANN_model = Sequential()

# Input + First Hidden Layer
ANN_model.add(Dense(
    32,
    activation='relu',
    input_dim=x_train.shape[1],
    kernel_initializer='he_uniform',
    kernel_regularizer=tf.keras.regularizers.l2(0.001)  # 🔥 FIXED
))
ANN_model.add(BatchNormalization())
ANN_model.add(Dropout(0.2))

# Hidden Layer
ANN_model.add(Dense(32, activation='relu'))
ANN_model.add(Dropout(0.2))

# Another Layer
ANN_model.add(Dense(16, activation='relu'))

# Output Layer
ANN_model.add(Dense(len(np.unique(y_train)), activation='softmax'))

# ==============================
# COMPILE
# ==============================
ANN_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Summary
ANN_model.summary()
```

```python
print(y_train.shape)
print(y_train[:5])
```

```python
ANN_history = ANN_model.fit(
    x_train,
    y_train,
    batch_size=32,
    epochs=500,
    validation_data=(x_test , y_test),
    callbacks=[callback]
)
```

```python
import numpy as np
import pandas as pd

# ==============================
# 1. NEW INPUT
# ==============================
new_data = pd.DataFrame([{
    'cation_exchange_capacity': 25,
    'humidity': 71,
    'nitrogen': 75,
    'phosphorus': 49.0,
    'potassium': 15.0,
    'rainfall': 102.083234,
    'soil_moisture': 6.0,
    'soil_ph': 5.902834,
    'temperature': 21.978982
}])

# ==============================
# 2. FEATURE ORDER
# ==============================
new_data = new_data[x_train.columns]

# ==============================
# 3. SCALING (SAFE)
# ==============================
try:
    new_data_scaled = scaler.transform(new_data)
except:
    new_data_scaled = new_data

# ==============================
# 4. ANN PREDICTION
# ==============================
ann_prob = ANN_model.predict(new_data_scaled)
ann_class = np.argmax(ann_prob, axis=1)[0]
ann_conf = np.max(ann_prob)

# ==============================
# 5. GB PREDICTION
# ==============================
gb_prob = gb_model.predict_proba(new_data)
gb_class = np.argmax(gb_prob, axis=1)[0]
gb_conf = np.max(gb_prob)

# ==============================
# 6. FUSION
# ==============================
w_ann = 0.3
w_gb  = 0.7

fusion_prob = (w_ann * ann_prob) + (w_gb * gb_prob)
fusion_class = np.argmax(fusion_prob, axis=1)[0]
fusion_conf = np.max(fusion_prob)

# ==============================
# 7. DECODE (SAFE)
# ==============================
def decode(label):
    try:
        return label.inverse_transform([label])[0]
    except:
        return f"Class_{label}"

ann_crop = decode(ann_class)
gb_crop = decode(gb_class)
fusion_crop = decode(fusion_class)

# ==============================
# 8. OUTPUT
# ==============================
print("----- PREDICTION RESULTS -----\n")

print("ANN Prediction      :", ann_crop, "| Confidence:", ann_conf)
print("Gradient Boosting   :", gb_crop,  "| Confidence:", gb_conf)
print("Fusion (Final)      :", fusion_crop, "| Confidence:", fusion_conf)
```
