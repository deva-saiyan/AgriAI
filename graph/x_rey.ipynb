```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = ['Custom CNN', 'ResNet-50', 'Fusion Model']

# Classes
classes = ['Normal', 'Tuberculosis', 'Pneumonia']

# Precision scores for each class
custom_cnn = [0.92, 0.93 ,0.94]
resnet50 = [0.95, 0.96, 0.97]
fusion_model = [0.97, 0.98, 0.99]

# X-axis positions
x = np.arange(len(classes))
width = 0.25

# Create figure
plt.figure(figsize=(10,6))

# Bar plots
plt.bar(x - width, custom_cnn, width, label='Custom CNN', color='#1E3A8A')
plt.bar(x, resnet50, width, label='ResNet-50', color='#4DA6FF')
plt.bar(x + width, fusion_model, width, label='Fusion Model', color='#D9D9D9')

# Labels and title
plt.xlabel('Classes')
plt.ylabel('Precision Score')
plt.title('Precision Comparison for Each Class')

# X-axis class names
plt.xticks(x, classes)

# Y-axis limit
plt.ylim(0.85, 1.00)

# Legend
plt.legend()

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = ['Custom CNN', 'ResNet-50', 'Fusion Model']

# Values
epochs = [17, 13, 21]
accuracy = [93.41, 96.12, 97.53]
precision = [92.87, 95.74, 97.18]
loss1 = [0.1823, 0.1103, 0.0712]
loss2 = [0.1974, 0.1289, 0.0843]

x = np.arange(len(models))
width = 0.25

# Unique colors per model
colors = ['#1E3A8A', '#4DA6FF', '#D9D9D9']





plt.figure(figsize=(10,6))

# Accuracy bars
plt.bar(x - width, accuracy, width, label='Accuracy (%)', color=colors[0])

# Precision bars
plt.bar(x, precision, width, label='Precision (%)', color=colors[1])

# Loss bars (scaled for visibility)
plt.bar(x + width, np.array(loss1)*100, width, label='Loss (Train x100)', color=colors[2], alpha=0.6)

# Title & labels
plt.title('Performance Comparison of Models')
plt.xlabel('Models')
plt.ylabel('Score')

plt.xticks(x, models)

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Value annotations
for i in range(len(models)):
    plt.text(x[i] - width, accuracy[i] + 0.3, f'{accuracy[i]}', ha='center', fontsize=9)
    plt.text(x[i], precision[i] + 0.3, f'{precision[i]}', ha='center', fontsize=9)
    plt.text(x[i] + width, loss1[i]*100 + 0.3, f'{loss1[i]}', ha='center', fontsize=9)

plt.legend()
plt.show()
```
