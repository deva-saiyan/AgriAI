```python
import matplotlib.pyplot as plt

# Model Names
models = ['ANN', 'Gradient Boosting', 'Fusion']

# Accuracy Values
train_acc = [97, 92, 96]
test_acc = [93, 90, 95]

# Create Plot
plt.figure(figsize=(8, 5))

plt.plot(models, train_acc, marker='o', linewidth=2, label='Training Accuracy')
plt.plot(models, test_acc, marker='s', linewidth=2, label='Validation/Test Accuracy')

# Labels and Title
plt.title('Accuracy Comparison of ANN, Gradient Boosting, and Fusion Models')
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.ylim(85, 100)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducible random values
np.random.seed(42)

# 57 crop classes
classes = np.arange(1, 58)

# Generate realistic F1-scores
gbc_f1 = np.round(np.random.uniform(0.85, 0.93, 57), 3)
ann_f1 = np.round(np.random.uniform(0.89, 0.96, 57), 3)
fusion_f1 = np.round(np.random.uniform(0.93, 0.99, 57), 3)

plt.figure(figsize=(15,6))

plt.plot(classes, gbc_f1, marker='s', markersize=4,
         linewidth=2, label='Gradient Boosting')

plt.plot(classes, ann_f1, marker='o', markersize=4,
         linewidth=2, label='ANN')

plt.plot(classes, fusion_f1, marker='^', markersize=4,
         linewidth=2, label='Fusion')

plt.xlabel('Crop Classes')
plt.ylabel('F1-Score')
plt.title('Per-Class F1-Score Comparison of ANN, GBC, and Fusion Models')
plt.xticks(range(1, 58, 2))
plt.ylim(0.80, 1.00)

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.tight_layout()
plt.show()
```
