```python
import matplotlib.pyplot as plt
import numpy as np

# 13 Pest Classes
classes = [
    "Stem Borer", "Leaf Roller", "Rice Bug", "Armyworm",
    "Leafhopper", "Gall Midge", "Thrips", "Hispa",
    "Cutworm", "Weevil", "Aphid", "Whitefly", "Locust"
]

# YOLOv8 Accuracy
yolo = [88, 90, 87, 89, 91, 88, 90, 87, 89, 90, 88, 91, 89]

# MobileNetV3Small Accuracy
mobile = [91, 92, 90, 91, 93, 90, 92, 89, 91, 92, 90, 93, 91]

# Fusion Model Accuracy
fusion = [94, 95, 93, 94, 96, 94, 95, 93, 94, 95, 94, 96, 95]

# X-axis
x = np.arange(len(classes))

# Figure
plt.figure(figsize=(14,6))

# Curves
plt.plot(x, yolo, marker='o', linewidth=2, label='YOLOv8', color='red')
plt.plot(x, mobile, marker='s', linewidth=2, label='MobileNetV3Small', color='blue')
plt.plot(x, fusion, marker='^', linewidth=2, label='Fusion Model', color='green')

# Labels
plt.xlabel("Pest Classes", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.title("Comparison of Accuracy Across 13 Pest Classes", fontsize=14)

# X ticks
plt.xticks(x, classes, rotation=25)

# Y-axis range
plt.ylim(80, 100)

# Grid
plt.grid(True, linestyle='--', alpha=0.5)

# Legend
plt.legend()

# Tight layout
plt.tight_layout()

# Show plot
plt.show()
```
